Cosine Similarity Algorithm

두 벡터 간 각도에 기반하여 구할 수 있는 알고리즘
벡터의 각도 방향에 따라 값을 반환함. 정반대의 경우 -1 반환cos.PNG

```python
from numpy import dot
from numpy.linalg import norm
import numpy as np

def cos(a,b):
  return dot(a,b) / (norm(a)*norm(b))


ar = np.array([1,1,1,0])
ar2 = np.array([0,0,1,1])
print(cos(ar,ar2))
```

유클리드 거리
다차원 공간에서의 점과 점을 구하는 공식
피타고라스 정리에서 삼각형 빗면 구하는 방식과 동일ucli.PNG

```python
def euclidean_dist(x,y):   
  return np.sqrt(np.sum((x-y)**2))
```

자카드 유사도 
각 집합에 대해 같은 원소를 얼마나 포함하고 있는 지를 통하여 찾는 기법
자카드 유사도는 0과 1사이 값을 가짐

a와 b의 합집합/교집합을 통하여 구할 수 있음.

```python
def jaccard(a,b):
  union=set(a).union(set(b))
  intersection = set(a).intersection(set(b))
  return len(intersection)/len(union)
```

```python
set1 = ['안녕', '오늘은', '아침을', '먹고', '산에서', '운동을', '할꺼야']
set2 = ['안녕', '오늘은', '점심을', '먹고', '바다에서', '운동을','할꺼야']
print(jaccard(set1,set2))
```
