```python
import numpy as np
import pandas as pd
from numpy import dot
from numpy.linalg import norm
import urllib.request
from sentence_transformers import SentenceTransformer
```

```python
import urllib.request
```

```python
train_data=pd.read_csv('당뇨 질의문 답변.csv',encoding='utf-8')
train_data2=pd.read_csv('일반 대화 데이터셋.csv',encoding='utf-8')
```

```python
train_data2=train_data2.dropna(axis=1)
```

```python
train_data.head()
train_data2[:20]
```

```python
t_data=pd.concat([train_data,train_data2],ignore_index=True)
```

```python
t_data
```

```python
model = SentenceTransformer('sentence-transformers/xlm-r-100langs-bert-base-nli-stsb-mean-tokens')
```

```python
t_data['embedding'] = t_data.apply(lambda row: model.encode(row.Qestion), axis=1)
```

```python
t_data
```

```python
def cos_sim(A, B):
  return dot(A, B)/(norm(A)*norm(B))
```

```python
def return_answer(question):
  embedding = model.encode(question)
  t_data['score'] = t_data.apply(lambda x: cos_sim(x['embedding'], embedding), axis=1)
  return t_data.loc[t_data['score'].idxmax()]['Answer']
```

```python
return_answer("당뇨에 좋은 음식 알려줘")
```
